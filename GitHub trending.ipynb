```
import urllib.request, json 
from phraseg import *
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import collections
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```
def read_url_data(open_url):
  try:
    with urllib.request.urlopen(open_url) as url:
      return url.read().decode()
  except:
    return ""
```

```
daily_trending = json.loads(read_url_data("https://github-trending-api.now.sh/repositories?&since=daily"))
print(daily_trending)
```

```
trending = ""
for t in daily_trending:
  url = t['url'].replace('https://github.com/','https://raw.githubusercontent.com/')
  url += "/master/README.md"
  print(url)
  trending += read_url_data(url)
print("Finish fetching repos")
```

```
phraseg = Phraseg(trending)
result = phraseg.extract()
```

```
print(result)
```

```
result = collections.OrderedDict(result)
result = {k: v for k, v in result.items() if len(list(k)) > 2 }
```

```
print(result)
```

```
wordcloud = WordCloud(font_path='./NotoSansCJKtc-Medium.otf', width=1800, height=1000, margin=2,background_color="white").fit_words(result)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
